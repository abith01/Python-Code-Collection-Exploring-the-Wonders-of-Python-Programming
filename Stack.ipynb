```python
class node():
    def __init__(self,data):
        self.data=data
        self.prev=None 
        self.next=None 
class dll():
    def __init__(self):
        self.head=None
        
#to an element in the beginning

    def addbeg(self,data):
        newnode=node(data)
        temp=self.head
        newnode.next=temp
        self.head=newnode
        
# to an element in the end
    def addend(self,data):
        newnode=node(data)
        temp=self.head
        while temp.next:
            temp=temp.next
        temp.next=newnode
        newnode.prev=temp
        
# Insert at the first position, respectively
    def addmid(self,data,loc):
        newnode=node(data)
        temp=self.head
        for i in range(0,loc-1):
            temp=temp.next
        newnode.prev=temp
        newnode.next=temp.next
        temp.next.prev=newnode
        temp.next=newnode
# Delete at the beginning
    def delbig(self):
        temp=self.head
        self.head=temp.next
        temp.next=None
        self.head.prev=None
        
# Delete at the end
    def delend(self):
        temp=self.head.next
        before=self.head
        while temp.next:
            temp=temp.next
            before=before.next
        
        before.next=None
        temp.prev=None
# delete in the specific location
    def delloc(self,loc):
        temp=self.head.next
        before=self.head
        for i in range(1,loc-1):
            temp=temp.next
            before=before.next
        before.next=temp.next
        temp.next.prev=before
        temp.next=None
        before.prev=None

    def display(self):
        if self.head is None:
            print("Empty node")
        else:
            temp=self.head
            while temp :
                print(temp.data,"<---->",end=" ")
                temp=temp.next
                
n1=node(10)
s1=dll()
s1.head=n1
n2=node(20)
n1.next=n2
n2.prev=n1.next
n3=node(30)
n2.next=n3
n3.prev=n2.next
n4=node(40)
n3.next=n4
n4.prev=n3.next
n5=node(50)
n4.next=n5
n5.prev=n4.next
s1.addbeg(5)
s1.addend(60)
s1.addmid(7,1)
s1.delbig()
s1.delend()
s1.delloc(3)
s1.display()
```
